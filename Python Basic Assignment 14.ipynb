### 1. What does RGBA stand for?

RGBA stands for Red, Green, Blue and Alpha. value of each RGB color is set separately and then collectively it will display the color of the computer screen. Alpha value denotes transparency of the image.

### 3. What is a box tuple, and how does it work?

Box tuple is a tuple with four integer value that represent the four coordinates of a square in any image in the sequence of Left, Top, Right, Bottom.

### 5. What method would you call to get Image object for a 100100 image, excluding the lower-left quarter of it?

```python
#lets resize the above image to 100 x 100 first 
small_image =imag.resize ((int(620-520),int (464-364)))
```
